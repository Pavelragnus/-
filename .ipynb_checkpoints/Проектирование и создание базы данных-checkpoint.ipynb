# Описание предметной области

**В качестве продементой области выбраны данные, применяемые в практической работе биоифнорматика - таблица с пациентами и аанотациями к имеющимся у них мутациям - к примеру, однонуклеотидным заменам.**   
**Таблица может быть посвящена исследованию пациентов с определенным заболеванием - в данном случае, болезнь Паркинсона.**    
**Для демонстрации выбраны мутации, относящиеся только к 4 хромосоме.**


**Приведем пример**

```python
import pandas as pd
data = pd.read_csv('Files/chr4.csv')
data.sample(5)
```

**Как видим таблица содержит первую колонку с описанием кодового имени пациента, а затем такие позиции, как:**    
- 'CHROM': номер хромосомы, в которой обнаружена мутация
- 'POS': локус мутации (расположение)
- 'ID': обозначение однонуклеотидной замены (точечной мутации), принятое в международных базах данных
- 'SYMBOL': символ гена, с которым ассоциирована мутация
- 'Gene': номер гена (ID), принятый в международных базах данных
- 'HGVSc': один из вариантов обозначения мутации
- 'gnomADg_AF_NFE': частота встречаемости этой мутации в популяции по данным базы gnomAD
- 'HGVSp': обозначение мутации по другому классификатору с указанием аминокислотной замены в кодируемом белке
- 'CADD_PHRED': специальный плагин, позволяющий увидеть степень опасности мутации (принято, что индекс более 20 указывает на высокую вероятность "вредности" мутации (deleteriousness)
- 'ClinVar_CLNSIG': указывает на клинические последствия мутации (патогенная, вероятно патогенная, доброкачесвтенная и так далее)
- 'ClinVar_CLNDN': указывает на конкретные заболевания, с которыми может быть ассоциирована мутация
- 'LoFtool':  еще один инструмент, позволяющий установить вероятность пагубности мутации
- 'CAROL': аналогично

**Подготовим таблицу для создания базы данных**:
- присвоим первой колонке название Participant
- уберем из обозначения пациентов расширение '.csv', оставшееся после предыдущих этапов анализа и аннотирования мутаций

```python
data = data.rename(columns={"Unnamed: 0": "Participant"})
data.sample(2)
```

```python
import re
data["Participant"] = data["Participant"].str.replace(".csv", "", regex=True)
data.sample(2)
```

# Создание структуры хранения данных в SQL

```python
import sqlite3
connect = sqlite3.connect("annotation_db")
c = connect.cursor()
```

```python

c.execute("CREATE TABLE IF NOT EXISTS Participants (Participant text, ID text)")
connect.commit()
```

```python
data[['Participant', 'ID']].to_sql("Participants", connect, if_exists="append", index=False)
```

```python
c.execute("CREATE TABLE IF NOT EXISTS Variants (ID text, POS text, SYMBOL text, HGVSc text, gnomADg_AF_NFE text, HGVSp text, CADD_PHRED float, ClinVar_CLNSIG text, ClinVar_CLNDN text, LoFtool float, CAROL text)")
connect.commit()
```

```python
data2 = data[['ID', 'POS', 'SYMBOL', 'HGVSc', 'gnomADg_AF_NFE', 'HGVSp', 'CADD_PHRED', 'ClinVar_CLNSIG', 'ClinVar_CLNDN', 'LoFtool', 'CAROL']].drop_duplicates()
```

```python
data2.to_sql("Variants", connect, if_exists="append", index=False)
```

```python
c.execute("CREATE TABLE IF NOT EXISTS Genes (SYMBOL text, Gene text, CHROM text)")
connect.commit()
```

```python
data_to_sql = data[['SYMBOL', 'Gene', 'CHROM']].drop_duplicates()
```

```python
data_to_sql.to_sql("Genes", connect, if_exists="append", index=False)
```

# Реализация запросов

## Получить список всех пациентов в базе (выведем первые 15 пациентов)

```python
c.execute(
    """  
    SELECT Participant from Participants
    LIMIT 15;
    """
)
```

```python

for row in c.fetchall():
    print(row)
```

## Получить список всех мутантных генов в базе (первые 15)

```python
c.execute(
    """  
    SELECT SYMBOL  from Genes
    LIMIT 15;
    """
)
```

```python

for row in c.fetchall():
    print(row)
```

## Какие заболевания могут быть у пациента с имеющимися у него мутациями? (например, мутации в гене AGA)

```python
c.execute(
    """  
    SELECT DISTINCT ClinVar_CLNDN FROM Variants
    WHERE SYMBOL='AGA';
    """
)
```

```python
for row in c.fetchall():
    print(row)
```

## Какой CADD-PHRED у конкретной мутации? (например, NP_620775.2:p.Thr141Ile по классификации HGVSp)

```python
c.execute(
    """  
    SELECT DISTINCT CADD_PHRED FROM Variants
    WHERE HGVSp='NP_620775.2:p.Thr141Ile';
    """
)
```

```python
for row in c.fetchall():
    print(row)
```

## Покажите пациентов с мутациями, имющими по CADD_PHRED высокий риск "вредности" (индекс более 20)? Выведем первые 15

```python
c.execute(
    """  
    SELECT DISTINCT Participant FROM Participants
    INNER JOIN Variants ON
    Participants.ID=Variants.ID
    WHERE CADD_PHRED >=20
    LIMIT 15;
    """
)
```

```python

for row in c.fetchall():
    print(row)
```

## Какому гену относится конкретная мутация?

```python
c.execute(
    """  
    SELECT DISTINCT SYMBOL FROM Variants
    WHERE HGVSp='NP_620775.2:p.Thr141Ile';
    """
)
```

```python
for row in c.fetchall():
    print(row)
```
